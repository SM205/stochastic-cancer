```python
pantherhuman = pd.read_csv("clean_pantherhuman.csv",index_col="family_id")
pantherhuman = pantherhuman[pantherhuman.index.value_counts() > 3]
overview = pd.DataFrame(pantherhuman[~pantherhuman.index.duplicated()]["family_name"])
overview.loc["Dataset"] = "Identifier"
```

```python
overview
```

```python
for dataset in main_folder_names:
    x = dataset+"Bootstrap/"
    for files in os.listdir(x):
        if os.path.isdir(os.path.join(x,files)):
            new = os.path.join(x,files)
            for file in os.listdir(new):
                bootstrap = pd.read_excel(new+"/Final_IC_bootstrap.xlsx")
                bootstrap = bootstrap[bootstrap["upper"] < 1]
                bootstrap = bootstrap[(bootstrap["non_zero_genes"] > 3) & (bootstrap["mean_gene_per_cell"] > 0.1)]
                exclusive = list(bootstrap["family_id"])
                exclusive_count = pd.Series(0, index=overview.index)
                for family in overview.index:
                    if family in exclusive:
                        exclusive_count[family] += 1
                exclusive_count["Dataset"] = dataset[9:-1]
                overview[files] = exclusive_count
```
