```python
exclusive_expression_df = pd.DataFrame()
for dataset in main_folder_names:
    x = dataset+"Bootstrap/"
    for files in os.listdir(x):
        if os.path.isdir(os.path.join(x,files)):
            new = os.path.join(x,files)
            dichotomised = pd.read_csv(new+"/clean_pantherhuman/intermediate/dichotomised_genes.csv",index_col='gene_id')
            familyset = pd.read_csv(new+"/clean_pantherhuman/intermediate/familyset.csv",index_col='gene_id')
            fam_list = exclusive_including_individual_enriched
            num_expressed = []
            for families in fam_list:
                genes = familyset[familyset['family_id'] == families].index
                expression = dichotomised.loc[genes].sum(axis=1)
                num_expressed.append(expression)
            exclusive_expression = pd.concat(num_expressed)
            exclusive_expression_df = pd.concat([exclusive_expression_df,exclusive_expression],axis=1)
            exclusive_expression_df.rename(columns={0:files},inplace=True)
```
